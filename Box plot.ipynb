# Box Plot

Box plots are figures that illustrate five descriptive statistics of numeric distributions: median, 25th percentile or Q1, 75th percentile or Q3, Q1 - 1.5*IQR (distribution min), and Q3 + 1.5*IQR (distribution max), with IQR being the interquartile range and is Q3 - Q1. 

In general, a box plot looks like the below illustration. 50% of the data is always within the box area, and the majority of data should be within the distribution min and max of which anything outside is considered outliers.



## Demonstration

### Load data

We use the `auto-mpg` data to demonstrate usages of box plots

```python
import pandas as pd

data = pd.read_csv('auto-mpg.csv')
data.head(2)
```

```python
data.info()
```

## Basic box plots

### Matplotlib

In matplotlib, we can draw box plots with `pyplot.boxplot()`. At minimum, we need to provide the input array.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.boxplot(data['mpg'])
plt.show()
```

We can set title with `pyplot.title()` and replace the tick label with `pyplot.xticks()` (the first input is actual values, second input their labels)

```python
column = 'mpg'
title = 'box plot of MPG'

plt.figure(figsize=(5,5))
plt.boxplot(data[column])
plt.xticks([1],[column])
plt.title(title)
plt.show()
```

We can customize the styles of all components in a box plot like the box, whisker (the vertical bar), cap (the distribution min/max), the median bar, and outliers (fliers)

```python
column = 'mpg'
title = 'box plot of MPG'

plt.figure(figsize=(5,5))
plt.boxplot(data[column],
            boxprops={'linewidth':4, 'color':'steelblue'},
            whiskerprops={'linewidth':4, 'color':'slateblue'},
            capprops={'linewidth':4, 'color':'blue'},
            medianprops={'linewidth':5, 'color':'red', 'alpha':0.5},
            flierprops={'marker':'H', 'markersize':'15', 'markerfacecolor':'orange'},
            widths=0.3)
plt.xticks([1],[column])
plt.title(title)
plt.show()
```

The mean can be showed as a point in the box plot as well by setting `showmean=True`. The style of the mean can be changed with `meanprops`

```python
column = 'mpg'
title = 'box plot of MPG'

plt.figure(figsize=(5,5))
plt.boxplot(data[column],
            showmeans=True,
            boxprops={'linewidth':4, 'color':'steelblue'},
            whiskerprops={'linewidth':4, 'color':'slateblue'},
            capprops={'linewidth':4, 'color':'blue'},
            medianprops={'linewidth':5, 'color':'red', 'alpha':0.5},
            meanprops={'marker':'D', 'markersize':10, 'markerfacecolor':'green'},
            flierprops={'marker':'H', 'markersize':'10', 'markerfacecolor':'orange'},
            widths=0.3)
plt.xticks([1],[column])
plt.title(title)
plt.show()
```

Filling the box with a color is slightly more complicated. We need set `patch_artist=True` in `boxplot()`

```python
column = 'mpg'
title = 'box plot of MPG'

plt.figure(figsize=(5,5))
plt.boxplot(data[column],
            showmeans=True,
            patch_artist=True,
            boxprops={'linewidth':4, 'facecolor':'lightblue', 'color':'steelblue'},
            whiskerprops={'linewidth':4, 'color':'slateblue'},
            capprops={'linewidth':4, 'color':'blue'},
            medianprops={'linewidth':5, 'color':'red', 'alpha':0.5},
            meanprops={'marker':'D', 'markersize':10, 'markerfacecolor':'green'},
            flierprops={'marker':'H', 'markersize':'10', 'markerfacecolor':'orange'},
            widths=0.3)
plt.xticks([1],[column])
plt.title(title)
plt.show()
```

### In pandas

In pandas, we call `plot.box()` from a series to draw its box plot

```python
data['mpg'].plot.box()
plt.show()
```

Like in matplotlib, we can adjust the styles of all components of the box plot, for example

```python
column = 'mpg'
title = 'box plot of MPG'

data[column].plot.box(
    figsize=(5,5),
    title=title,
    showmeans=True,
    patch_artist=True,
    boxprops={'linewidth':4, 'facecolor':'lightblue', 'color':'steelblue'},
    whiskerprops={'linewidth':5, 'color':'steelblue'},
    capprops={'linewidth':5, 'color':'steelblue'},
    medianprops={'linewidth':5, 'color':'red', 'alpha':0.5},
    meanprops={'marker':'D', 'markersize':10, 'markerfacecolor':'green'},
    flierprops={'marker':'o', 'markersize':'10', 'markerfacecolor':'orange'},
    widths=0.3
)
plt.show()
```

### In seaborn

In seaborn, we use `boxplot()`. Like matplotlib, at minimum, we need to provide the array

```python
import seaborn as sn

sn.boxplot(data['mpg'])
plt.show()
```

seaborn has some of its own options to stylize the box plot

```python
column = 'mpg'
title = 'box plot of MPG'

plt.figure(figsize=(5,5))
sn.boxplot(
    data[column],
    showmeans=True,
    linewidth=4,
    color='lightblue',
    width=0.3)
plt.xticks([0],[column])
plt.title(title)
plt.show()
```

Most of the same props options in matplotlib are still usable if you want more customization

```python
column = 'mpg'
title = 'box plot of MPG'

plt.figure(figsize=(5,5))
sn.boxplot(
    data[column],
    showmeans=True,
    color='lightblue',
    boxprops={'linewidth':5, 'edgecolor':'steelblue'},
    whiskerprops={'linewidth':5, 'color':'steelblue'},
    capprops={'linewidth':5, 'color':'steelblue'},
    medianprops={'linewidth':5, 'color':'red', 'alpha':0.5},
    meanprops={'marker':'D', 'markersize':10, 'markerfacecolor':'purple'},
    flierprops={'marker':'o', 'markersize':'10', 'markerfacecolor':'orange'},
    width=0.3)
plt.xticks([0],[column])
plt.title(title)
plt.show()
```

## Multiple box plots

We may want to draw multiple box plots to compare their distributions.

### In matplotlib

Simply feed a 2D array or a (numeric) DataFrame to `pyplot.boxplot()` to draw one plot for each column. We need to manually feed the xtick labels and rotate them if needed

```python
num_cols = ['acceleration', 'cylinders', 'displacement', 'horsepower', 'mpg', 'weight']
title = 'box plots of numeric columns'

plt.figure(figsize=(5,5))
plt.boxplot(data[num_cols])
plt.xticks(range(1,7), num_cols, rotation=30)
plt.title(title)
plt.show()
```

We can scale the columns so the plots look better. While we lose the information on the columns' range, we keep the shape of their distributions. All props options are still available

```python
num_cols = ['acceleration', 'cylinders', 'displacement', 'horsepower', 'mpg', 'weight']
title = 'box plots of standardized numeric columns'

plt.figure(figsize=(10,5))
data_scaled = (data[num_cols] - data[num_cols].mean())/data[num_cols].std()
plt.boxplot(
    data_scaled,
    boxprops={'linewidth':4, 'color':'steelblue'},
    whiskerprops={'linewidth':4, 'color':'steelblue'},
    capprops={'linewidth':4, 'color':'steelblue'},
    medianprops={'linewidth':5, 'color':'red', 'alpha':0.5},
    flierprops={'marker':'o', 'markersize':'10', 'markerfacecolor':'orange'},
    widths=0.5
    )
plt.xticks(range(1,7), num_cols, rotation=30)
plt.title(title)
plt.show()
```

### In pandas

In pandas, we can either draw all box plots in the same figure with `DataFrame.plot.box()`, or separately in subfigures.

## Stratified box plots

box plots can be stratified by a categorical column to investigate the differences in distributions of data across the classes. 

While this can be done in matplotlib, it is much more complicated than pandas or seaborn, so let's just skip to the latter two.

### In pandas

We first slice to the two columns of interests then call `boxplot()` with preferred styles. The stratifying column is provided with `by=`

```python
#set columns of interest
column = 'mpg'
strat_column = 'origin'

#draw box plots
data[[column,strat_column]].boxplot(
    by=strat_column, 
    rot=20, 
    boxprops={'linewidth':4, 'color':'steelblue'},
    whiskerprops={'linewidth':4, 'color':'steelblue'},
    capprops={'linewidth':4, 'color':'steelblue'},
    medianprops={'linewidth':5, 'color':'red', 'alpha':0.5},
    flierprops={'marker':'o', 'markersize':'10', 'markerfacecolor':'orange'},
    figsize=(6,6))
plt.show()
```

### In seaborn

In seaborn we can set the stratifying column with either `x=` or `y=` which makes the plots vertical or horizontal, respectively

```python
column = 'mpg'
strat_column = 'origin'
title = 'box plots of MPG grouped by Origin'

plt.figure(figsize=(6,6))
sn.boxplot(
    data,
    x=strat_column,
    y=column,
    showmeans=True,
    linewidth=4,
    flierprops={'marker':'o', 'markersize':'10', 'markerfacecolor':'orange'},
    meanprops={'marker':'D', 'markersize':10, 'markerfacecolor':'lightgray'}
)
plt.xticks(rotation=30)
plt.title(title)
plt.show()
```
